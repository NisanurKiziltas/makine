```python
#maaaş ve yaş arasındaki ilişki
from pickle import PicklingError


df1 = df[['age','wage']]

#girdi -> input, feature -> x
x=df1['age']

#çıktı -> output,label -> y
y=df1['wage']

#ortalama hesapla
y_mean = df1.groupby('age').mean()

#grafiği hazırla
fig, ax = plt.subplots(figsize=(8,8))

#datayı çiz
ax.scatter(x,y, color='orange')

#ortalamayı çiz
ax.plot(y_mean.index,y_mean)

#grafiği formatla
plt.title('maaş-yaş')
plt.xlabel('yaş')
plt.ylabel('maaş')

plt.show()
```

```python
#maaş - yıl
from turtle import color


df2 = df[['year', 'wage']]

# girdi -> feature -> x
x = df2['year']

#çıktı -> output, label -> y
y = df2['wage']

#ortalamayı hesapla
y_mean = df2.groupby('year').mean()

#grafiği hazırla
fig, ax = plt.subplots(figsize=(8,8))

#datayı çiz
plt.scatter(x,y,color = 'pink')

#ortalamayı çiz
ax.plot(y_mean.index, y_mean, color='blue')

#grafiği formatla
plt.title('maaş - yıl')
plt.xlabel('yıl')
plt.ylabel('maaş')
plt.show()
```
